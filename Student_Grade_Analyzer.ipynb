```python
# ============================================
# Day 1 Project: Student Grade Analyzer
# Using NumPy
# ============================================

import numpy as np
scores = np.random.randint(0, 100, size=50)

print("========== STUDENT SCORES ==========")
print(scores)
mean_score = np.mean(scores)
median_score = np.median(scores)
highest_score = np.max(scores)
lowest_score = np.min(scores)
std_deviation = np.std(scores)

print("\n========== BASIC STATISTICS ==========")

print(f"Mean Score       : {mean_score:.2f}")
print(f"Median Score     : {median_score}")
print(f"Highest Score    : {highest_score}")
print(f"Lowest Score     : {lowest_score}")
print(f"Standard Deviation : {std_deviation:.2f}")
failed_students = scores[scores < 40]

print("\n========== FAILED STUDENTS (<40) ==========")
print(failed_students)

Excellent_students = scores[scores > 85]

print("\n========== Excellent STUDENTS (>85) ==========")
print(Excellent_students)


normalized_scores = (
    (scores - np.min(scores))
    /
    (np.max(scores) - np.min(scores))
)

print("\n========== NORMALIZED SCORES ==========")
print(normalized_scores)


reshaped_scores = scores.reshape(5, 10)

print("\n========== RESHAPED SCORES (5x10) ==========")
print(reshaped_scores)


row_averages = np.mean(reshaped_scores, axis=1)

print("\n========== ROW-WISE AVERAGES ==========")
print(row_averages)


percentile_25 = np.percentile(scores, 25)
percentile_75 = np.percentile(scores, 75)

print("\n========== PERCENTILES ==========")

print(f"25th Percentile : {percentile_25}")
print(f"75th Percentile : {percentile_75}")

theory_scores = scores
practical_scores = np.random.randint(0, 100, size=50)
final_scores = (0.4 * theory_scores) + (0.6 * practical_scores)

print("\n========== THEORY SCORES ==========")
print(theory_scores)

print("\n========== PRACTICAL SCORES ==========")
print(practical_scores)

print("\n========== FINAL WEIGHTED SCORES ==========")
print(final_scores)

topper_index = np.argmax(scores)
topper_score = np.max(scores)

print("\n========== TOPPER DETAILS ==========")

print(f"Topper Student Index : {topper_index}")
print(f"Topper Score         : {topper_score}")
```
